```python
import seaborn as sns
import numpy as np
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
X=iris.petal_length
Y=iris.species
Y.unique()
```

```python
idx=np.random.permutation(np.arange(Y.size))
nTrain=120
xTrain=X[idx[:nTrain]]
yTrain=Y[idx[:nTrain]]
xTest=X[idx[nTrain:]]
yTest=Y[idx[nTrain:]]
```

```python
yTest.size
```

```python
py=np.zeros(len(Y.unique()))
for i in range(len(py)):
  py[i]=(yTrain==Y.unique()[i]).sum()/yTrain.size
#calculating the probability of each class from Y
py
```

```python
fx=np.zeros(len(py)*2)
fx.resize(len(py),2)
for i in range(len(py)):
  a=xTrain[yTrain==Y.unique()[i]]
  fx[i,0]=a.mean()
  fx[i,1]=a.std()
fx
```

```python
yTest=np.asarray(yTest)
xTest=np.asarray(xTest)
```

```python
#FINAL MODEL



# xt=xTest[0]
xt=float(input("ENTER THE PETAL LENGTH: "))
guassValue=np.zeros(3)
for i in range(len(Y.unique())):
  mu=fx[i][0]
  sgma=fx[i][1]
  guassValue[i]=py[0]*(np.exp(-(0.5*(((xt-mu)/sgma)**2))))/(sgma*((2*np.pi)**0.5))
guassValue=list(guassValue)
final=guassValue.index(max(guassValue))
print("SPECIES:",Y.unique()[final])
```
